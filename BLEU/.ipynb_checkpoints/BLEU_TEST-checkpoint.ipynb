```python
from collections import Counter
import numpy as np
from nltk import ngrams
import nltk.translate.bleu_score as bleu
import openpyxl
```

```python
wb = openpyxl.load_workbook('BLEU_Score.xlsx')
sheet = [wb['Ensemble BLEU'], wb['Model 1 BLEU'], wb['Model 2 BLEU'], wb['Model 3 BLEU'], wb['Model 4 BLEU'], wb['Model 5 BLEU']]


for s in sheet:
    print(s)
    for i in range(2, 1002):
        print((i-1) , '번째')
        candidate = str(s.cell(row = i, column = 4))
        references = str(s.cell(row = i, column = 2))

        # NLTK 패키지 구현되어져 있는 코드로 계산한 BLEU 점수
        print(bleu.sentence_bleu(list(map(lambda ref: ref.split(), references)),candidate.split()))

        bleu_s = bleu.sentence_bleu(list(map(lambda ref: ref.split(), references)),candidate.split())
        s.cell(row = i, column = 5).value = bleu_s
    
wb.save('BLEU_Score.xlsx')
```
